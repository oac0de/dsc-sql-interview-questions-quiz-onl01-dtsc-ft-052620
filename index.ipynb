## Question 2

_______________________________________________________________________________________________________________________________

Primary Keys are unique identifiers in datatables. A Foreign key is that Primary Key in a different table. These two are used to relate or connect databases/datatables. 




## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
There can be one-to-one relationship, which uses a singular entity to join another singular entity. For example, one customer's firstname can retrieve that customer's one lastname. 

There can be a one-to-many relationship, where one entity retrieves multiple entities. For example, one customer's orderNumber can retrieve multiple products ordered.

There is also many-to-many relationships, where multiple entities can retrieve multiple entities. These are to be coded for carefully as a many-to-many relationship can possibly overload a system's database as the resulting datatable size can increase exponentially. For example, a table containing multiple offices to be joined with a table of employees per office, will multiply the two counts together.


## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
There are INNER JOINs, which SQL defaults as 'JOIN'. INNER JOINs are for singular Primary Keys or Foreign Keys to link matching values among two data sets.

There are LEFT JOINS which join all the values from one table, with only matching values from another table.

There are RIGHT JOINs join one table's values that are matching with another table, and all the values from that second table.

There are also OUTER JOINS, which only are possible in versions of SQL other than SQLite3, which joins all values/columns from both tables.
